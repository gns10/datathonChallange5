```python
colunasTrabalho = ['RA', 'IPS', 'IDA', 'IEG']

padraoAlunos2023 = pede2023[pede2023['RA'].isin(pede2022['RA'])]
padraoAlunos2024 = pede2024[pede2024['RA'].isin(pede2023['RA'])]
```

```python
padraoAlunos2022 = pede2022[colunasTrabalho]
padraoAlunos2023 = padraoAlunos2023[colunasTrabalho]
padraoAlunos2024 = padraoAlunos2024[colunasTrabalho]
```

```python
df = pd.concat([
    padraoAlunos2022.assign(ANO=2022),
    padraoAlunos2023.assign(ANO=2023),
    padraoAlunos2024.assign(ANO=2024)
])

df = df.sort_values(['RA', 'ANO'])
```

```python
df = df.dropna(subset=['IPS', 'IDA', 'IEG'])
df_filtrado = df[df.groupby('RA')['RA'].transform('count') > 1]
df_filtrado = df_filtrado.sort_values('RA')
df_filtrado.reset_index(drop=True, inplace=True)
```

```python
df_filtrado
```

```python
df_filtrado = df_filtrado.sort_values(['RA', 'ANO'])

df_filtrado['IDA_next'] = df_filtrado.groupby('RA')['IDA'].shift(-1)
df_filtrado['IEG_next'] = df_filtrado.groupby('RA')['IEG'].shift(-1)


# remover última observação de cada aluno
df_model = df_filtrado.dropna(subset=['IDA_next', 'IEG_next'])
```

```python
df_model
```

```python
df_model['diff_IDA'] = df_model['IDA_next'] - df_model['IDA']
df_model['diff_IEG'] = df_model['IEG_next'] - df_model['IEG']
```

```python
df_model
```

```python
diff = df_model.groupby('IPS')[['diff_IDA', 'diff_IEG']].mean().sort_values('diff_IDA')
print(diff)
```

```python
diff.plot(kind='bar')
plt.axhline(0, linestyle='--')
plt.title('Impacto do IPS na evolução do aluno')
plt.show()
```

```python
#padraoAlunos2022_7 = pede2022[colunasTrabalho7]
padraoAlunos2023_7 = pede2023[colunasTrabalho7]
padraoAlunos2024_7 = padraoAlunos2024_7[colunasTrabalho7]
```

```python
df = pd.concat([
    #padraoAlunos2022_7.assign(ANO=2022),
    padraoAlunos2023_7.assign(ANO=2023),
    padraoAlunos2024_7.assign(ANO=2024)
])

df = df.sort_values(['RA', 'ANO'])
```

```python
df['IDA_cat'] = pd.qcut(df['IDA'], 3, labels=['baixo', 'medio', 'alto'])
df['IEG_cat'] = pd.qcut(df['IEG'], 3, labels=['baixo', 'medio', 'alto'])
df['IPS_cat'] = pd.qcut(df['IPS'], 3, labels=['baixo', 'medio', 'alto'])
df['IPP_cat'] = pd.qcut(df['IPP'], 3, labels=['baixo', 'medio', 'alto'])
```

```python
from sklearn.cluster import KMeans

df_cluster = df[['IDA', 'IEG', 'IPS', 'IPP', 'INDE']].dropna().copy()

df_cluster['cluster'] = KMeans(n_clusters=4, random_state=42).fit_predict(
    df_cluster[['IDA', 'IEG', 'IPS', 'IPP']]
)

df_cluster.groupby('cluster')['INDE'].mean().sort_values(ascending=False)
```

```python
from sklearn.tree import DecisionTreeRegressor

df_tree = df[['IDA', 'IEG', 'IPS', 'IPP', 'INDE']].dropna().copy()

X = df_tree[['IDA', 'IEG', 'IPS', 'IPP']]
y = df_tree['INDE']

tree = DecisionTreeRegressor(max_depth=3, random_state=42)
tree.fit(X, y)
```

```python
from sklearn.tree import export_text

print(export_text(tree, feature_names=list(X.columns)))
```
